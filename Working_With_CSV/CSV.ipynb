```python
#Opening a CSV file from an URL
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0(Macintosh; Intel Mac Os x 10:14; rv:66:0) Gecko/20100101 Firfox/66.0"} 
req = requests.get(url,headers=headers)
data= StringIO(req.text)

pd.read_csv(data)
```

```python
#header parameter
pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv", header=1)
```

```python
#use_cols Parameters
pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv", usecols=['Country' , 'Region'])
```

```python
#Squeeze parameters
import pandas as pd
# Read the CSV and convert to Series using .squeeze() method instead of squeeze parameter
df = pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv", usecols=['Region']).squeeze()

# Alternative approach: Access the column directly to get a Series
# df = pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv", usecols=['Region'])['Region']
```

```python
#skiprows parameters
df = pd.read_csv(
    "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv",
    skiprows=1
)
```

```python
#Econding Parameter
df = pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv",encoding='latin-1')
```

```python
#Skip  bad Lines
df = pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv",encoding='latin-1', on_bad_lines='skip')
```

```python
#dtypes parameter
df = pd.read_csv(
    "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv",
    dtype={'Region':int}).info()
```

```python
#Handling Dates
df = pd.read_csv(
    "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv",
    parse_dates=['date']).info()
```

```python
#convertors
def rename(name):
    if name == "AFRICA":
        return "Africa"
    else:
        return name
```

```python
rename("AFRICA")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv", converters= {'Region': rename})
```

```python
#na_values parameters
pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv",na_values=["Botswana"])
```

```python
#Loading a hug dataset in chunks
dfs = pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv", chunksize=50)
```
