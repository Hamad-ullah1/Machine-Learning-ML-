```python
#if you have dataset use this:
#df = pd.read_csv('covid_toy.csv')

#if you don't have use this
import pandas as pd

url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day28-column-transformer/covid_toy.csv"

df = pd.read_csv(url)

df.head()
```

```python
#1)In Column 'gender' we use OneHotEncoder
#2)In Column 'fever' we use SimpleImputer
#3)In Column 'cough' we use OrdinalEncoder
#4)In Column 'city' we use OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test , y_train, y_test = train_test_split(
    df.drop(columns=['has_covid']),
    df['has_covid'], test_size=0.2)
```

```python
x_train
```

If You Don't Know , How to use Column Transformer , Use this

```python
#adding simple imputer to fever colunm
from sklearn.impute import SimpleImputer

si = SimpleImputer()

# train data
x_train_fever = si.fit_transform(x_train[['fever']])

# test data (IMPORTANT: only transform, not fit)
x_test_fever = si.transform(x_test[['fever']])

print(x_train_fever.shape)
```

```python
x_train_fever
```

```python
#Ordinalencoding -> cough
oe = OrdinalEncoder(categories=[['Mild','Strong',]])
x_train_cough = oe.fit_transform(x_train[['cough']])

#Also the test Data
x_test_cough = oe.fit_transform(x_test[['cough']])

print(x_train_cough.shape)
```

```python
x_train_cough
```

```python
#OneHotEncoder -> gender , city
oh = OneHotEncoder(drop='first', sparse_output=False)
x_train_gender_city = oh.fit_transform(x_train[['gender', 'city']])

#Also the test Data
x_test_gender_city = oh.transform(x_test[['gender', 'city']])

print(x_train_gender_city.shape)
```

```python
#Extracting age
x_train_age = x_train.drop(columns=['gender', 'fever', 'cough', 'city']).values

#Also the test Data
x_test_age = x_test.drop(columns=['gender', 'fever', 'cough', 'city']).values

x_train_age
```

```python
print(x_train_gender_city)
```

```python
x_train_transformer = np.concatenate(
    [x_train_age, x_train_fever, x_train_gender_city, x_train_cough],
    axis=1
)

x_test_transformer = np.concatenate(
    [x_test_age, x_test_fever, x_test_gender_city, x_test_cough],
    axis=1
)

x_train_transformer.shape
```

If You Know , How to use Column Transformer , So we Use this

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
    transformers=[
        ('t1', SimpleImputer(), ['fever']),
        ('t2', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough']),
        ('t3', OneHotEncoder(drop='first', sparse_output=False), ['gender', 'city'])
    ],
    remainder='passthrough'
)
```

```python
transformer.fit_transform(x_train)
```

```python
transformer.fit_transform(x_train).shape
```
