```python
#if percentage is possible
df["vote_count"].value_counts().plot(kind='pie', autopct='%.2f%%')
#its looks messy , so we use another way
```

```python
counts = [
    (df["vote_count"] > 1000).sum(),
    (df["vote_count"] <= 1000).sum()
]

labels = ["Above 1000", "1000 or Below"]

import matplotlib.pyplot as plt

plt.pie(counts, labels=labels, autopct="%1.1f%%")
plt.show()
```
