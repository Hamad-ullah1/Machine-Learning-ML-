```python
import numpy as np

import pandas as pd



from sklearn.model_selection import train_test_split

from sklearn.impute import SimpleImputer

from sklearn.preprocessing import OneHotEncoder

from sklearn.preprocessing import MinMaxScaler

from sklearn.tree import DecisionTreeClassifier
```

```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("hesh97/titanicdataset-traincsv")

print(path)

files = os.listdir(path)
print(files)
```

```python
df = pd.read_csv(os.path.join(path, files[0]))
df.head()
```

```python
df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin',],inplace=True)
```

```python
df.head()
```

```python
x_train, x_test, y_train , y_test =train_test_split(
    df.drop(columns=['Survived']),
            df['Survived'],
            test_size=0.2,
            random_state=42)
```

```python
x_train.head(2)
```

```python
y_train.head(2)
```

```python
df.isnull().sum()
```

```python
#Applying Imputation
from sklearn.impute import SimpleImputer

s_age = SimpleImputer()
s_embarked = SimpleImputer(strategy='most_frequent')

# Training
x_train_age = s_age.fit_transform(x_train[['Age']])
x_train_embarked = s_embarked.fit_transform(x_train[['Embarked']])

# Testing
x_test_age = s_age.transform(x_test[['Age']])
x_test_embarked = s_embarked.transform(x_test[['Embarked']])
```

```python
x_train_age
```

```python
x_test_embarked
```

```python
#One hot encoding Sax and Embarked
ohe_sex = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Training
x_train_sex = ohe_sex.fit_transform(x_train[['Sex']])
x_train_embarked = ohe_embarked.fit_transform(x_train_embarked)

# Testing
x_test_sex = ohe_sex.transform(x_test[['Sex']])
x_test_embarked = ohe_embarked.transform(x_test_embarked)
```

```python
x_train_sex
```

```python
x_train_embarked
```

```python
x_train.head(2)
```

```python
x_train_rem = x_train.drop(columns=['Sex', 'Age', 'Embarked'])
```

```python
x_test_rem = x_test.drop(columns=['Sex', 'Age', 'Embarked'])
```

```python
x_train_transformed = np.concatenate((x_train_rem , x_train_age ,
                                     x_train_sex , x_train_embarked) , axis=1)

x_test_transformed = np.concatenate((x_test_rem , x_test_age ,
                                     x_test_sex , x_test_embarked) , axis=1)
```

```python
x_train_transformed
```

```python
x_train_transformed.shape
```

```python
x_test_transformed.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_transformed , y_train)
```

```python
y_pred = clf.predict(x_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred)
```
