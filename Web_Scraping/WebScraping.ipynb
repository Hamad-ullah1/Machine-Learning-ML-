Fetching data using web Scraping

```python
#imports
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#If it is not working
#requests.get('https://www.ambitionbox.com/list-of-companies?page=1').text
```

```python
#use this. So the web thing that request comming from brower
final = pd.DataFrame()

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.3'
}
```

```python
# Loop Through Pages
for j in range(1, 11):

    url = f'https://www.ambitionbox.com/list-of-companies?page={j}'

    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')

    company = soup.find_all('div', class_='company-content-wrapper')

    name = []
    rating = []
    review = []
    ctype = []
    hq = []
    employees = []
```

```python
# Extract Data
for i in company:

        # Name
        name.append(i.find('h2').text.strip())

        # Rating
        rating.append(i.find('p', class_='rating').text.strip())

        # Reviews
        review.append(i.find('a', class_='review-count').text.strip())

        # Type
        ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())

        # Headquarters
        hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())

        # Employees
        employees.append(i.find_all('p', class_='infoEntity')[2].text.strip())
```

```python
#Create DataFrame
d = {
    'name': name,
    'rating': rating,
    'review': review,
    'type': ctype,
    'hq': hq,
    'employees': employees
}

df = pd.DataFrame(d)

final = pd.concat([final, df], ignore_index=True)
```
